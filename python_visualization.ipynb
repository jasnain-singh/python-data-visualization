```python
# Writing score vs math score
plt.figure (figsize=(10,6))
plt.scatter(df["math score"],df["writing score"],marker="*",color="y")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.title("Scatter plot [Math score VS writing score]")
```

# Data Visualization

```python
ds = pd.read_csv("gas_prices.csv")
ds
```

### Line Chart
1. Line chart of Year vs Australia Canada France (combined)
2. Line chart of Year vs Germany Italy Japan (combined)
3. Line chart of Year vs Mexico South Korea UK USA (combined)

```python
plt.figure(figsize=(12,8))
plt.plot(ds["Year"],ds["Australia"], marker="2", color="r", label="Australia")
plt.plot(ds["Year"],ds["Canada"], marker=".", color="c", label="Canada")
plt.plot(ds["Year"],ds["France"], marker="o", color="y", label="France")
plt.xlabel("Year")
plt.ylabel("Australia - Canada - France")
plt.title("Year vs Australia Canada France")
plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.plot(ds["Year"],ds["Germany"], marker="2", color="r", label="Germany")
plt.plot(ds["Year"],ds["Italy"], marker=".", color="c", label="Italy")
plt.plot(ds["Year"],ds["Japan"], marker="o", color="y", label="Japan")
plt.xlabel("Year")
plt.ylabel("Germany - Italy - Japan")
plt.title("Year vs Germany Italy Japan")
plt.legend()
```

```python
plt.figure(figsize=(12,8))
plt.plot(ds["Year"],ds["Mexico"], marker="2", color="r", label="Mexico")
plt.plot(ds["Year"],ds["South Korea"], marker=".", color="c", label="South Korea")
plt.plot(ds["Year"],ds["UK"], marker="s", color="y", label="UK")
plt.plot(ds["Year"],ds["USA"], marker="o", color="b", label="USA")
plt.xlabel("Year")
plt.ylabel("Mexico - South Korea - UK - USA")
plt.title("Year vs Mexico South Korea UK USA")
plt.legend()
```

```python
# changing the figure size
# changing the front size of the figure

plt.figure(figsize=(10,8))
plt.plot(ds.Year, ds.USA, '*-', label='USA', color='r')
plt.plot(ds.Year, ds.Australia, '*-', label='Australia', color='g')
plt.plot(ds.Year, ds.UK, '*-', label='UK', color='b')
plt.plot(ds.Year, ds.Japan, '*-', label='Japan', color='y')
#marker = *, line = -, color = red

plt.title('Gas Prices over Year(USD)', fontdict={'fontweight':'bold', 'fontsize':18})
# changing the title font size and font type

plt.xlabel('Year', fontdict={'fontweight':'bold', 'fontsize':12})
plt.ylabel('Price', fontdict={'fontweight':'bold', 'fontsize':12})
plt.legend()
#saving the image in particular format(png,jpg) resolution dpi == 100

plt.savefig('Gas Prices over Year(USD).png', dpi=100)
```
